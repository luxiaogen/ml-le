```python
from sklearn.linear_model import SGDRegressor
```

```python
# 1. 定义数据
# 自变量，每周学习时长
X = [[5], [8], [10], [12], [15], [3], [7], [9], [14], [6]]
# 因变量，数学考试成绩
y = [55, 65, 70, 75, 85, 50, 60, 72, 80, 58]
```

```python
# 2. 定义模型
model = SGDRegressor(
    penalty=None,
    loss='squared_error',    # 损失函数
    max_iter=10**6,  # 最大迭代次数
    eta0=1e-5,       # 学习率
    learning_rate='constant',   # 常数学习率
    tol=1e-8    # 停止迭代策略
)

# 3. 模型训练
model.fit(X, y)
```

```python
print(model.coef_)
print(model.intercept_)
```
